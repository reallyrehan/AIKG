```python


def addColor(p):
    txt = str(p)
    color = "black"
    
    if p.is_keyword or isinstance(p,sqlparse.sql.Where):    
        color = "blue"
        txt = txt.upper() 
    else:
        if str(p)=="*" or isinstance(p,sqlparse.sql.Identifier):
            color="black"
        else:
            color = "red"

        
    
    return ' <span style = "color:'+color+';">'+txt+'</span>'

def sqlHighlighter(query):
    ans = ""
    parsed = sqlparse.parse(query)[0]
    
    for p in parsed.tokens:
        print(type(p))

        ans = ans +addColor(p)
    return ans
```
